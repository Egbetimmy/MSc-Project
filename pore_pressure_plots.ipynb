```python
import pandas as pd
from class2 import vp_to_vs
```

```python
data = pd.read_csv('data/well_4_ML.csv')
```

```python
data['vs'] = vp_to_vs(data, 'velocity')
```

```python
data
```

```python
from pore_pressure import eaton_method, Bowers_method, Matthews_Kelly_method
```

```python
data['Pore_pressure_1'] = eaton_method(data['velocity'], data['vs'], data['PHI'])
data
```

```python
data['Pore_pressure_2'] = Bowers_method(data['velocity'], data['vs'], data['PHI'])
data['Pore_pressure_3'] = Matthews_Kelly_method(data['DEPT'], data['DEN'])
```

```python
data.head()
```

```python
df = data
df
```
