```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
```

```python
df_sorted = df.sort_values('Depth')
```

```python
sns.set_style("whitegrid")
fig, ax1 = plt.subplots(figsize=(8, 12))

# Plot actual vs predicted values for training set
ax1.plot(df_sorted['Actual Permeability'], df_sorted['Depth'], label='Actual')
ax1.plot(df_sorted['gb_Permeability'], df_sorted['Depth'], label='Predicted')
ax1.set_xlabel('PERM')
ax1.set_ylabel('Depth')
ax1.set_title('Actual vs Predicted Values - Training Set')
ax1.set_xscale('log')
ax1.legend()

# Show the plot
plt.show()
```

```python
sns.set_style("whitegrid")
fig, ax1 = plt.subplots(figsize=(8, 12))

# Plot actual vs predicted values for training set
sns.scatterplot(x=df_sorted['Actual Permeability'], y=df_sorted['Depth'], color='black', label='Actual', ax=ax1)
sns.scatterplot(x=df_sorted['gb_Permeability'], y=df_sorted['Depth'], color='red', label='Predicted', ax=ax1)

ax1.set_xlabel('PERM')
ax1.set_ylabel('Depth')
ax1.set_title('Actual vs Predicted Values - Training Set')
ax1.set_xscale('log')
ax1.invert_yaxis()
ax1.legend()

# Show the plot
plt.show()
```

```python
# Create a scatter plot of the predicted and actual values in a log-log plot
sns.scatterplot(x='Actual Permeability', y='gb_Permeability', data=df_sorted, alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Actual Permeability')
plt.ylabel('Predicted Permeability')
plt.title('Actual vs. Predicted Permeability')
plt.show()
```
